```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
dataset2=pd.read_csv("spambase.data")
dataset=np.loadtxt("poker-hand-testing.data",delimiter=',')
# https://www.youtube.com/watch?v=99MN-rl8jGY
# S1,C1,S2,C2,S3,C3,S4,C4,S5,C5,CLASS
```

```python
dataset.head()
```

```python
print(dataset[0])
```

```python
x=dataset[:,0:10]
y=dataset[:,-1]
x
```

```python
X_train, X_test,y_train,y_test=train_test_split(x,y,test_size = .33, random_state=17)
```

```python
BenNB=BernoulliNB(binarize=True)
BenNB.fit(X_train,y_train)
y_expect=y_test
print(BenNB)
y_pred = BenNB.predict(X_test)
print(accuracy_score(y_expect,y_pred))
```

```python
MultiNB=MultinomialNB()
MultiNB.fit(X_train,y_train)
print(MultiNB)
y_pred=MultiNB.predict(X_test)
print(accuracy_score(y_expect,y_pred))
```

```python
GausNB=GaussianNB()
GausNB.fit(X_train,y_train)
print(GausNB)
y_pred=GausNB.predict(X_test)
print(accuracy_score(y_expect,y_pred))
```
